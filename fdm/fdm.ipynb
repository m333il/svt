```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def solve_function(name):
    size = [10, 25, 50, 100, 200, 250, 300, 400, 500, 600, 700, 800, 900, 1000]
    l2_norm = []
    c_norm = []
    for n in size:
        print(f'n = {n}')
        os.system("./build/fdm " + str(n))
        with open('norms.txt') as f:
            norms = f.readlines()[0].split()
            l2_norm.append(float(norms[0]))
            c_norm.append(float(norms[1]))
    plt.plot(size, l2_norm, label = "L2 norm")
    plt.plot(size, c_norm, label = "C norm")
    plt.legend()
    plt.grid()
    plt.yscale("log")
    plt.xlabel("Количество узлов")
    plt.title("Ошибка для задачи ")
    plt.savefig(name)
```

```python
solve_function('1.png')
```

```python
solve_function('2.png')
```
